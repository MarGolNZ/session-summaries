```python
mdata = pd.read_csv("student-mat.csv", sep= ";")
```

From the .describe() function I can see the mean, min and max values of all the classifications or features of the data. 
<br>
For instance, the mean age of the students taking Portugese (which is the total dataset) is 16.74, however of those students who also take Mathematics, the mean age is 16.69 which means that some older students decided to not take maths classes. 
<br>
<br>
Now to ask questions about the data
<br>
<br>
How many and to portion of the maths students are female?

```python
mdata["sex"].value_counts()
```

```python
mdata["sex"].value_counts(normalize=True)
```

What was the mean grade at their final maths exam?

```python
mdata["G3"].mean()
```

What was the mean grade at the final maths exam by gender?

```python
mdata[mdata["sex"]=="M"]["G3"].mean()
```

```python
mdata[mdata["sex"]=="F"]["G3"].mean()
```

What was the mean grade of the final maths exam for students who are 16?

```python

mdata[(mdata["age"]== 16)]["G3"].mean()
```

```python
mdata[(mdata["age"]== 16) & (mdata["sex"]=="F")]["G3"].mean()
```

```python
mdata[(mdata["age"]== 16) & (mdata["sex"]=="M")]["G3"].mean()
```

Now using the .groupby function, what was the grade of the final maths exam by gender and age?

```python
mdata.groupby(["sex","age"])["G3"].agg(['mean'])
```

And just out of curiosity, what was the mean grade of students in romantic relationships compared to those not in romantic relationships; and does their grade get impacted based on how many hours they spent studying every week?

```python
mdata.groupby(["romantic","studytime"])["G3"].agg(['mean','count'])
```
